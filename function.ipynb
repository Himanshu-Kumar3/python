```
class Point :
  def __init__(self , x , y):
    self.x = x
    self.y = y
  def __str__(self):
    return f"({self.x} , {self.y})"

  def distance(self , other):
    x1 = self.x
    x2 = other.x
    y1 = self.y
    y2 = other.y
    return ((x2-x1)**2 + (y2-y1)**2)**0.5

  def distance_from_origin(self):
    return (self.x**2 + self.y**2)**0.5


# Class for line
class Line :
  def __init__(self  , A , B , C):

        # line Formula = Ax + By + C = 0
    self.A = A
    self.B = B
    self.C = C
  def __str__(self):
    return f" {self.A}x + {self.B}y + {self.C} = 0"

  def point_in_line(self , p1):
    x = p1.x
    y = p1.y
    if self.A*x + self.B*y + self.C ==0:
      return f"point {p1} lies in line {self}"
    else:
      return f"point {p1} doesn't lies in line {self}"
  # distance formula -> d = abs(Ax + By + c) / sqrt(A**2 + B**2)
  def distance_from_point_to_line(self,p1):
    x = p1.x
    y = p1.y

    d = abs(self.A*x + self.B*y + self.C)/(self.A**2 + self.B**2)**0.5
    return d


p1 =Point(3,4)
p2 = Point(3,3)
print(p1,p2)
p1.distance(p2)
p1.distance_from_origin()

L1 = Line(2,3,4)
print(L1)
print(L1.point_in_line(p1))

L2 = Line(3 , 4, -5)
p3 = Point(1,2)
p3.z = 21
print(p3.x , p3.y , p3.z)

print(L2.distance_from_point_to_line(p3))
```

# Super Keyword
 - Special keyword by which child can access parent's constructor as well as method of the parent
  - can't access the data / variable
  - only be used inside the child

```
# Example :- smartPhone and phone

class Phone :
  def __init__(self , price , brand , camera):
    print("Parent Constructor")
    self.price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
    print(f"bought phone in {self.price}")
class SmartPhone(Phone) :
  def __init__(self , name , price , brand , camera):
    print("Child's constructor")
    super().__init__(price , brand , camera)
    self.name = name
    print("Child's constructor")
  def buy(self):
    super().buy()

smartPhone = SmartPhone("IQ009" ,15000 ,"IQ","50mpx" )

print(smartPhone.price,
smartPhone.brand,
smartPhone.camera,
smartPhone.name)

smartPhone.buy()

```

```
# Types of Inheritence :-
# - Single inheritence :- child -> inherits from parent
# - mutilevel -> child -> inherits from parent -> inherits from grand parent
# - hierarchical -> child1  , child2 , child3 ... -> inherits from parent
# - hybrid -> combination of many type of inheritence
# - multiple -> child -> inherits from parent1 , parent2 -> not possible in java (due to ambiguity) ->possible in pyhton ->

# Mutiple example
class Father :
  def __init__(self , father):
    self.father = father
    print(f"father constructor: {father}")

class Mother:
  def __init__(self , mother):
    self.mother = mother
    print(f"Mother constructor: {mother}")

# child class inherits the first , the constructor of that class will be executed
class Child(Father , Mother):
  def __init__(self, father_name, mother_name):
    # When using multiple inheritance with super(), it's crucial to understand MRO.
    # For distinct arguments like here, it's often clearer to call parents' __init__ explicitly.
    Father.__init__(self, father_name) # Explicitly call Father's constructor
    Mother.__init__(self, mother_name) # Explicitly call Mother's constructor
    print("Child constructor executed")

  def __str__(self):
    return "I am Child"


# Now, when creating a Child object, provide arguments for both parent constructors
ch1 = Child("John", "Jane")
print(ch1 , ch1.father , ch1.mother)

```

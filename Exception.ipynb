# Exception Handling
  - in python that allows to handle the errors gracefully and take corrective measures / actions without affecting / stopping the execution of the program

  - handles try , except , else  , finally


 - **Exception**
    - event that disrupt the normal flow of program execution .
    - Occurs when the error is encountered during the program executionn

```

# Examples of exception handling
# a = b  # NameError : to handle this
try :
  a=b
except NameError as ex:
  print(ex)
  # print("the value of b is not defined")  # after this the code is executed perfectly without disrupting the execution
```

```
try :
  n = int(input("enter the division"))
  num = 3/n
  a = b
except NameError as ex:
  print(ex)
except ZeroDivisionError as ex1:
  print(ex1)
```

```
# setting a main error that handle all kind of error
try :
  n = int(input("enter the division"))
  num = 3/n
  a = b
except ZeroDivisionError as ex1:
  print(ex1)
except Exception as ex2 :
  print(ex2)
  print('Main error')

```

```
# use of else :- this block is executed only if the code inside the try block doesn't returns any error
try :
  n = int(input("enter the division"))
  result= 3/n
except ZeroDivisionError as ex1:
  print(ex1)
except Exception as ex2 :
  print(ex2)
  print('Main error')
else:
  print(f"result is  {result}")  # only executes when the try block is correct
```

```
# the use of finally :- executes whether it returns error or not :- in every situation
try :
  n = int(input("enter the division"))
  result = 3/n
except ZeroDivisionError as ex1:
  print(ex1)
except Exception as ex2 :
  print(ex2)
  print('Main error')
else:
  print(f"the final result is  {result}")
finally:
  print("The code execution is completed")  # executed in every case whether try returns error or not
```
